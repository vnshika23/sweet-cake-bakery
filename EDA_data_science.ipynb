```python
#  Performing EDA (Exploratory Data Analysis) on coffee sales dataset

import pandas as pd
import matplotlib.pyplot as plt
# Load Dataset
import seaborn as sns

df = pd.read_csv("coffe_sales_data.csv")

print("\nColumns name:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())

print("\nShape of dataset:")
print(df.shape)

print("\nDataset Info:")
df.info()

print("\nMissing values:")
print(df.isnull().sum())

print("\nStatistical summary:")
print(df.describe())

# Total and Average Sales
print("\nTotal Sales:", df["money"].sum())
print("Average sales:", df["money"].mean())


# Univariate Analysis
# Sales distribution
plt.figure()
plt.hist(df["money"], bins=20)
plt.title("Sales Distribution")
plt.xlabel("Sales (Money)")
plt.ylabel("Frequency")
plt.show()


# Hours of day distribution
plt.figure()
plt.hist(df["hour_of_day"], bins=24)
plt.title("Hour of Day Distribution")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.show()


# Coffee count
plt.figure()
df["coffee_name"].value_counts().plot(kind="bar")
plt.title("Coffee Count")
plt.xlabel("Coffee Name")
plt.ylabel("Count")
plt.show()


# Bivariate Analysis
# Coffee vs Sales
plt.figure()
df.groupby("coffee_name")["money"].sum().plot(kind="bar")
plt.title("Coffee vs Total Sales")
plt.xlabel("Coffee Name")
plt.ylabel("Sales")
plt.show()


# Time of Day vs Sales
plt.figure()
df.groupby("Time_of_Day")["money"].mean().plot(kind="bar")
plt.title("Time of Day vs Average Sales")
plt.xlabel("Time of Day")
plt.ylabel("Average sales")
plt.show()


# Weekday vs Sales
plt.figure()
df.groupby("Weekday")["money"].mean().plot(kind="bar")
plt.title("Weekday vs Average sales")
plt.xlabel("Weekday")
plt.ylabel("Average sales")
plt.show()


# Correlation heatmap
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Multivariate Analysis: Correlation Heatmap")
plt.show()


# Data Analysis
df["Date"] = pd.to_datetime(df["Date"])

plt.figure()
df.groupby(df["Date"].dt.month)["money"].sum().plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
